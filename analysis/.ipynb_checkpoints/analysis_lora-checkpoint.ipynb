```python
def compare_layer_types(model):
    """比较三类LoRA层的特性差异"""
    results = {'attn.c_attn': [], 'attn.c_proj': [], 'mlp.c_proj': []}
    
    for layer_idx in range(24):  # GPT-2 medium有24层
        for layer_type in results.keys():
            module = model.base_model.transformer.h[layer_idx]
            # 获取对应模块
            if layer_type == 'mlp.c_proj':
                target = module.mlp.c_proj
            else:
                target = module.attn.get_submodule(layer_type.split('.')[1])
            
            delta_w = target.lora.weight
            # 计算关键指标
            sv = torch.linalg.svdvals(delta_w)
            effective_rank = (sv > 0.01 * sv[0]).sum().item()
            sparsity = (delta_w.abs() < 1e-4).float().mean().item()
            
            results[layer_type].append({
                'effective_rank': effective_rank,
                'sparsity': sparsity,
                'norm': delta_w.norm().item()
            })
    
    # 分析：MLP层的适配器是否与注意力层表现不同？
    # 如果是，可能解释了为何BLEU没提升
    return results
```
